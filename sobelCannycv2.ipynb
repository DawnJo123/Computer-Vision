```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread("sample1.jpg", cv2.IMREAD_GRAYSCALE)  # Read in grayscale

# Sobel Edge Detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in x-direction
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in y-direction
sobel_combined = cv2.magnitude(sobel_x, sobel_y)       # Magnitude of gradients

# Canny Edge Detection
canny_edges = cv2.Canny(image, 100, 200)  # Thresholds: 100 (low), 200 (high)

# Plot the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.title("Sobel - Combined")
plt.imshow(sobel_combined, cmap="gray")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.title("Sobel - X Direction")
plt.imshow(np.abs(sobel_x), cmap="gray")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.title("Canny Edges")
plt.imshow(canny_edges, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
