```python
import cv2
import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("traffic.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
pixels = image.reshape(-1, 3)  # Flatten the image into RGB vectors

# Apply GMM
gmm = GaussianMixture(n_components=2, covariance_type='tied', random_state=0)
gmm.fit(pixels)
labels = gmm.predict(pixels)

# Reshape labels to the image shape
segmented_image = labels.reshape(image.shape[:2])

# Display results
import matplotlib.pyplot as plt

# Assuming 'segmented_image' is your image data
plt.imshow(segmented_image)
plt.axis('off')
plt.title("Segmented Image")
plt.show()
```
